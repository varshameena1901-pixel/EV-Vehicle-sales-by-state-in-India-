```python
import pandas as pd
df = pd.read_csv("Electric Vehicle Sales by State in India.csv")
print(df.head())
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

```python
# Clean & preprocess data
df['Year'] = df['Year'].astype('int')
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
categorical_cols = ['Month_Name','State','Vehicle_Class','Vehicle_Category','Vehicle_Type']
df[categorical_cols] = df[categorical_cols].astype('category')
```

```python
### Perfrom Exploratory Data Analysis
```

```python
#A. Yearly EV Sales Trend
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))
sns.lineplot(x='Year', y='EV_Sales_Quantity', data=df , marker='o')
plt.title('Yearly EV Sales in India')
plt.show()
```

## EV sales increased slowly at first and then grow sharply after 2019.
2022-2023 show the highest sales.

```python
##B = Monthly Trend
sns.lineplot(x='Month_Name', y='EV_Sales_Quantity', data=df , marker='o')
```

## Sales remain fairly stable across month with only small fluctuations.

```python
##C. State-wise EV Sales
plt.figure(figsize=(10,12))
sns.barplot(y='State',x='EV_Sales_Quantity',data=df)
```

## Maharashtra, karnataka and Uttar Pradesh have the highest EV sales, while small states show lower adoption.

```python
##D. Vehicle Class Analsis
sns.barplot(x='Vehicle_Class',y='EV_Sales_Quantity',data=df)
plt.xticks(rotation=90)
```

## Moter cars, Scooters and Goods Carriers contribute most of EV demand.

```python
##Vehicle Category Analysis
sns.barplot(x='Vehicle_Category',y='EV_Sales_Quantity',data=df)
```

## 2-Wheelers and 3-Wheelers dominate EV sales in india.

```python
## Vehicle Type Analysis
sns.barplot(x='Vehicle_Type',y='EV_Sales_Quantity',data=df)
plt.xticks(rotation=90)
```

## 2W_Personal is the most popular EV type;
## 3W Shared and Goods types follow next.

```python
## Advanced Analysis
```

```python
## Top 10 States by EV Sales
df.groupby('State')["EV_Sales_Quantity"].sum().sort_values(ascending=False).head(10)
```

```python
## Which vehicle type sell the most
df.groupby('Vehicle_Type')["EV_Sales_Quantity"].sum().sort_values(ascending=False)
```

```python
## Year on year growth calculation
df.groupby("Year")["EV_Sales_Quantity"].sum().pct_change()
```

## EV adoption in india is rising fast, especially after 2019.
## top states lead the growth, and 2-wheelers make up most of the EV market.
